## 1. Handling Missing Values

### Check for missing values

```python
df.isnull().sum()
```

### Currently no missing value if have Drop rows with missing values

```python
df.dropna(inplace=True)
df
```

### Fill missing values with mean/median/mode

```python
df['Income'] = df['Income'].fillna(df['Income'].mean())
df['Employment_Status'] = df['Employment_Status'].fillna(df['Employment_Status'].mode().iloc[0])
df
```

### Drop duplicates (if any duplicate raised, then proceed to drop duplicates)

## 3. Data Type Corrections

### Check data types

```python
df.dtypes
```

### Convert columns to correct types

```python
# Convert Loan_Term to integer
df['Loan_Term'] = df['Loan_Term'].astype(int)

# Convert Interest_Rate to numeric, set invalid values to NaN()int or float)(errors='coerce'
#If a value cannot be converted to a number (like a string "N/A" or "unknown"), it will be turned into NaN (Not a Number) instead of raising an error.)
df['Interest_Rate'] = pd.to_numeric(df['Interest_Rate'], errors='coerce')

# Example: Convert integer column to string
# df['Loan_ID'] = df['Loan_ID'].astype(str)

df
```

## 4 IQR (Interquartile Range)

```python
ex:
[100, 120, 130, 150, 160, 170, 180, 200, 1000]
In this case, 1000 is way above the upper bound, so it's considered an outlier and removed.

This helps improve the accuracy of statistical analysis and machine learning models.
```

```python
#IQR detects outliers
#below the lower bound or above the upper bound is considered an outlier. effective even for skewed data.
#(Skewness measures how asymmetrical a data distribution is compared to a perfectly symmetrical (normal) distribution.)
Q1 = df['Loan_Amount'].quantile(0.25)
Q3 = df['Loan_Amount'].quantile(0.75)
IQR = Q3 - Q1
df = df[(df['Loan_Amount'] >= Q1 - 1.5*IQR) & (df['Loan_Amount'] <= Q3 + 1.5*IQR)]
df
```

### 5. Confidence Intervals

#### Using a 95% confidence interval for mean Loan_Amount:

```python
import scipy.stats as stats
import numpy as np

data = df['Loan_Amount'].dropna()
mean = np.mean(data)
std = np.std(data, ddof=1)
n = len(data)
ci = stats.t.interval(0.95, n-1, loc=mean, scale=std/np.sqrt(n))
ci
```

### 6. Hypothesis Testing

#### "Is the average Loan_Amount greater than 200?"

```python
stats.ttest_1samp(df['Loan_Amount'].dropna(), 200)
```

### 7. ANOVA

#### Example: Compare mean Applicant_Income across Marital_Status

```python
from scipy.stats import f_oneway

single = df[df['Marital_Status']=='Single']['Income']
married = df[df['Marital_Status']=='Married']['Income']

f_oneway(single, married)
```
